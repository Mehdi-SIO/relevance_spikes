# Load dataset into raw_dataset_source

```python
raw_dataset = []
raw_dataset_source = []
with open('octo_intranet.json', 'r') as text_file:
    for line in text_file:
        item = json.loads(line)
        raw_dataset.append(item)
        raw_dataset_source.append(item['_source'])
```

```python
raw_preview_contents = []
for preview in raw_dataset_source:
    content = preview['preview_screenshot_content'].replace('\'', ' ') \
        .replace('`', ' ')\
        .replace('.',' ')\
        .replace(',', ' ')\
        .replace('\n', ' ')\
        .replace('/', ' ')\
        .replace('(', ' ')\
        .replace(')', ' ')
    content = content.lower()
    content = re.sub(' +', ' ',content)
    raw_preview_contents.append(content)
```

# Draw a word cloud on all the document corpus

```python
stop_words = get_stop_words('french')
stop_words.append('étre')
stop_words.append('d`un')
stop_words_set = Set(stop_words)
```

```python
word_count = {}
for preview_content in raw_preview_contents:
    words = preview_content.split(' ')
    for word in words:
        if len(word) > 3 and word not in stop_words_set:
            if word not in word_count:
                word_count[word]=1
            else:
                word_count[word]+=1
                
wordcloud = WordCloud(background_color="white", max_words=100).generate_from_frequencies(word_count.items())
plt.imshow(wordcloud)
plt.axis("off")
```

```python
def cloudword_by_text(lookup, stopwords_complements = ''):
    stop_words = get_stop_words('french')
    stop_words.append('étre')
    stop_words.append('d`un')
    stop_words_set = Set(stop_words) | Set(stopwords_complements) | Set(get_stop_words('en'))
    
    word_count_local={}
    for preview_content in raw_preview_contents:
        if lookup in preview_content:
            words = preview_content.split(' ')
            for word in words:
                if len(word) > 3 and word not in stop_words_set:
                    if word not in word_count_local:
                        word_count_local[word]=1
                    else:
                        word_count_local[word]+=1
                        
    #word_count_local_items=sorted(word_count_local.items(), key=itemgetter(1), reverse=True)
    #print(word_count_local_items[0:200])
    wordcloud = WordCloud(background_color="white", max_words=100).generate_from_frequencies(word_count_local.items())
    plt.imshow(wordcloud)
    plt.axis("off")
```

```python
cloudword_by_text('hadoop')
```

```python
cloudword_by_text('devops')
```
